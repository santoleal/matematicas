## 3. Operaciones Aritméticas y Matemáticas Básicas

### Descripción:
##### En Python, se pueden realizar operaciones básicas como suma, resta, multiplicación y división. También se pueden utilizar funciones matemáticas como potencia, raíz cuadrada, y más.

### Código:

```python
import math

# Operaciones Aritméticas
x = 5
y = 3

# Suma, resta, multiplicación y división
print(f"{x} + {y} = {x + y}")
print(f"{x} - {y} = {x - y}")
print(f"{x} * {y} = {x * y}")
print(f"{x} / {y} = {x / y}")

# Funciones Matemáticas
print(f"Raíz cuadrada de {x} es: {math.sqrt(x)}")
print(f"{x} elevado a la {y} es: {math.pow(x, y)}")
```

## 4. Estructuras de Datos

## 5. Control de Flujo

### Descripción:
##### El control de flujo nos permite tomar decisiones (condicionales) y repetir acciones (bucles). Esto es útil para resolver problemas matemáticos de forma iterativa o condicional.

### Código:

```python
# Condicionales
edad = 20
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")

# Bucle for
for i in range(1, 6):
    print(f"El número {i} al cuadrado es {i**2}")
```

## 6. Funciones Matemáticas

### Descripción:
##### Las funciones nos permiten encapsular lógica que se repite. Podemos definir funciones matemáticas que realicen operaciones específicas y reutilizarlas a lo largo del código.

### Código:

```python
# Definir una función para calcular el área de un círculo
def area_circulo(radio):
    return math.pi * radio ** 2

# Llamar a la función
radio = 5
print(f"El área del círculo con radio {radio} es: {area_circulo(radio)}")
```

## 7. Manejo de Archivos

### Descripción:
##### Podemos usar Python para leer y escribir datos en archivos. Esto es útil cuando tenemos grandes cantidades de datos para trabajar, como una lista de números que necesitan ser analizados.

### Código:

```python
# Escribir en un archivo
with open('numeros.txt', 'w') as file:
    file.write("1\n2\n3\n4\n5\n")

# Leer desde un archivo
with open('numeros.txt', 'r') as file:
    lineas = file.readlines()
    numeros = [int(line.strip()) for line in lineas]

# Calcular la suma de los números leídos
print(f"La suma de los números es: {sum(numeros)}")
```

## 8. Módulos Matemáticos

### Descripción:
##### Python tiene módulos integrados como math y librerías externas como NumPy para realizar operaciones matemáticas más avanzadas. Vamos a ver cómo usarlos.

### Código:

## 9. Visualización de Datos Matemáticos

### Descripción:
##### Con la librería Matplotlib, podemos graficar funciones matemáticas y datos, lo que nos permite visualizar resultados de cálculos o relaciones matemáticas de manera gráfica.

### Código:

```python
import matplotlib.pyplot as plt
import numpy as np

# Graficar una parábola
x = np.linspace(-10, 10, 100)
y = x**2

plt.plot(x, y, label="y = x^2")
plt.title("Gráfico de una parábola")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

## 10. Programación Orientada a Objetos (POO) aplicada a Matemáticas

### Descripción:
##### La Programación Orientada a Objetos nos permite crear clases y objetos para representar conceptos matemáticos. Por ejemplo, podemos modelar un círculo como una clase con propiedades como el radio y métodos como el cálculo del área.

### Código:

```python
# Definir una clase para un Círculo
class Circulo:
    def __init__(self, radio):
        self.radio = radio
    
    def area(self):
        return math.pi * self.radio ** 2

# Crear un objeto de la clase Circulo
mi_circulo = Circulo(5)
print(f"El área del círculo es: {mi_circulo.area()}")
```

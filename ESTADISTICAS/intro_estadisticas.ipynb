### <u>Conceptos</u>

* **ESTADÍSTICA**: Ante todo, es una ciencia. Eso quiere decir que se rige por métodos y procedimientos científicos. Su objetivo es **recopilar, organizar, analizar e interpretar datos de la fenómenos naturales o sociales (realidad)**, orientándose a la toma de decisiones fundamentadas. 

    - Es una de rama de las matemáticas, pero es la rama más cercana a las ciencias sociales. O sea, aplica herramientas de análisis matemáticos para obtener información de la realidad concreta y dinámica.

    - A diferencia de la probabilidad (que es puramente teórica), la estadística se basa en DATOS, y desde allí se sustentan las evidencias.

    - Se basa en métodos y técnicas para extraer información significativa y tomar decisiones fundamentadas. Estas deben ser escogidas de acuerdo a naturaleza del problema.


* **PROBABILIDAD**: Es una medida matemática. Permite cuantificar la incertidumbre asociada a un evento. Fluctúa entre el valor "0" (imposible) y valor "1" (seguro). A diferencia de la estadística (que si bien están profundamente ligados, se diferencian), la probabilidad es puramente teórica. Considera definiciones como: 
    - Experimento aleatorio (resultado no puede anticiparse del todo; determinado por el azar).
    - Espacio muestral (conjunto de todos los posibles resultados).
    - Evento / suceso (subconjunto de resultados posibles).


* **ESTADÍSTICAS DESCRIPTIVAS**: Describe y resume datos observados. Responde al qué, al cuándo, al dónde. Busca caractetizar e identificación de propiedades de un conjunto de datos observados. Algunos de sus métodos son:
    - Tablas de frecuencias y gráficos.
    - Medidas de tendencia central (media o promedio, mediana y moda).
    - Medidas de dispersión (varianza, rango, desviación estándar).
    - Indicadores de posición (separación en partes iguales).    

* **ESTADÍSTICA INFERENCIAL**: Analiza una muestra de la población total, y generaliza sus conclusiones (inferencias). Se asocia a enfoques inductivos (de la muestra a la caracterización de la población), a diferencia de enfoque deductivo (que reduce desde lo general, como lo es la est. descriptiva). Algunos de sus métodos son:
    - Estimación de parámetros.
    - Prueba de hipótesis: determina si hay o no evidencias suficientes para aceptar o rechazar hipótesis.
    - Intervalos de confianza.

* **ESTADÍSTICA PREDICTIVA** (por complementar): Utiliza modelos econométricos como:
    - Regresión lineal
    - Regresión logística


* **DATOS**: Registro de la realidad. Tienen diferente origen, y permiten hacer referencias al tipo de fenómeno observado: geográfico, astronómico, médico, cultural, meteorológico, académico, financiero, gubernamental, comunitario, laboral, etc.
    - Se dice que los datos, después de ser procesados y analizados, pasan a ser **INFORMACIÓN**, y esta, cuando se transmite y activa reacción/interacción, pasa a ser **COMUNICACIÓN**.



* **TIPOS DE DATOS**: Forma de categorizar en base a la naturaleza y propiedades de los datos. Su identificación permite definir el tipo de análisis (análisis de variables cualitativas o análisis de variables cuantitativos). Hay cuatro tipos:
    
    - Categóricos (cualitativos):
        - Nominales: Categorización o clasificación de elementos sin orden ni jerarquía. La operación es solo de igualdad/desigualdad (coincidencias).
            - Ejemplos: Apellidos, nacionalidad, género, color de ojos...  
        - Ordinales: Mantención de orden o jerarquía, pero distancia no es cuantificable. Sus operaciones son de igualdad/desigualdad, y mayor/menor.
            - Ejemplos: talla de ropa (S/M/L/XL), satisfacción, nivel educacional... 

    - Numéricos (cuantitativos):
        - De intervalo: Son datos numéricos que tienen un orden y una distancia cuantificable entre las categorías. No hay punto cero absoluto (no parten de cero). Sumas y restas sí, pero no la división. 
            - Ejemplos: puntajes PSU, Temperatura en Celcius, fecha... 
        - De razón: Son datos numéricos que tienen orden, distancia cuantificable y un punto cero absoluto. Todas las operaciones tienen sentido.
            - Ejemplos: Renta mensual, distancia, edad, peso...


* **OTROS TIPOS DE DATOS NUMÉRICOS O CUANTITATIVOS**: A los tipos de datos de intervalo o razon, se suman otras formas de caracterizar y complementar. Son fundamentales para el análisis de la naturaleza de los datos.

    - Discretos: Se pueden contar con las manos. Soin finitos. Son valores enteros, contables y aislados.
        - Ejemplos: Edad, número de estudiantes, goles de un partido, tarjetas de jueces...
   
    - Continuos: Son medibles como decimal, por lo que pueden tomar cualquier valor de un rango. Su valore posible es infito, representa números reales. Precisión depende de instrumento con que se mide.
        - Ejemplos: Largo de una mesa,temperatura, promedio de notas.

* **REGISTRO DE DATOS**: Es una forma de comprender e indicar las fuentes y formas de organización de los datos a nivel computacional. Estos son:
    - Estructurados: tablas, base de datos SQL, hojas de cálculo tipo Excel.
    - Semiestructurados: XML, JSON, APIs, correos electrónicos.
    - No estructurados: informes, imágenes, audio, videos, literatura, bases de datos No-SQL...

* **ANÁLISIS EXPLORATORIO DE DATOS (EDA)**: Concepto utilzado para una revisión inicial de un conjunto de datos. 
    - Incrementa el conocimiento de los datos.
    - Distribuciones: visualiza la estructura de datos.
    - Identifica variables importantes.
    - Detecta anomalías/outliers.
    - Verifica suposiciones iniciales.
    - Genera hipótesis desde los datos.
    - Tipo de análisis depende de tipo de datos.

* **DATASET**: Es una colección o conjunto de datos, que generalmente está tabulada. O sea, serían estructurada en su origen, o bien posterior a su análisis (como puede ser tras el conteo de palabras de un libro). Hay algunos conceptos claves:
    - Objeto: Es una fila de la tabla (una observación, una "nueva entrada").
    - Atributo: Es una columna de la tabla (una variable, una categoría).
    - Los datos serían los valores que asumen los atributos.




#### <u>**3 conceptos operativos básicos de las estadísticas y probabilidades**</u>

* **POBLACIÓN**: Conjunto completo de todos los elementos que comparten algún atributo o característica en común, y que se quiere analizar.

* **MUESTRA**: Es un subconjunto representativo de población, seleccionado para el análisis que se quiere realizar.

* **VARIABLE**: Característica o propiedad que puede tomar diversos valores, de acuerdo a los elementos.

Entonces:

    De una **POBLACIÓN** determinada se extrae una **MUESTRA** y se miden **VARIABLES**. Todo ello se analiza con **MÉTODOS ESTADÍSTICOS**, aplicando un análisis cuantitativo o cualitativo según los **TIPOS DE DATOS** que se identifiquen.

##### <u>**¿Se puede mentir con estadísticas (o sea manipular números o interpretaciones)?**</u>

   - Todo depende de definiciones, encuadres, profundidades de análisis.
   - No hay análisis objetivos, siempre hay una selección de variables.
   - Los distintos descriptivos dan nociones distintas sobre los datos.


Ejemplo: jugador de fútbol.

- Estadísticas descriptivas: resumir el historiual deportivo.
- Estadística inferencial: predecir el desempeño futuro del jugador.
- _¿Cuál es el mejor jugador y de mejor proyección?_ Depende de lo que se entienda como mejor jugador, los datos considerados, la proyección temporal que se considera...
